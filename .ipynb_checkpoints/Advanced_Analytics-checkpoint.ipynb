```python
import pandas as pd

df = pd.read_csv("data/processed/nav_history_clean.csv")

df["date"] = pd.to_datetime(df["date"])

results = []

for code in df["amfi_code"].unique():
    fund = df[df["amfi_code"] == code].copy()

    fund["daily_return"] = fund["nav"].pct_change()

    var95 = fund["daily_return"].quantile(0.05)

    cvar95 = fund[fund["daily_return"] <= var95]["daily_return"].mean()

    results.append([code, var95, cvar95])

report = pd.DataFrame(
    results,
    columns=["amfi_code", "VaR_95", "CVaR_95"]
)

report.to_csv(
    "data/processed/var_cvar_report.csv",
    index=False
)

print(report.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("data/processed/nav_history_clean.csv")

key_funds = df["amfi_code"].unique()[:5]

for code in key_funds:

    fund = df[df["amfi_code"] == code].copy()

    fund["return"] = fund["nav"].pct_change()

    rolling_sharpe = (
        fund["return"].rolling(90).mean()
        /
        fund["return"].rolling(90).std()
    ) * np.sqrt(252)

    plt.plot(
        fund.index,
        rolling_sharpe,
        label=str(code)
    )

plt.legend()

plt.title("Rolling 90 Day Sharpe")

plt.savefig(
    "reports/rolling_sharpe_chart.png"
)

plt.show()
```
